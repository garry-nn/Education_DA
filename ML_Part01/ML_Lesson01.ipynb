## **1. Принцип машинного обучения**

---

Машинное обучение похоже на обучение живых существ. Живое существо, в том числе человек, учатся на своем опыте. Программа обучается на данных, которые она может сама загружать из онлайн-источников и специальных репозиториев.

- Чем больше опыт человека, тем лучше он решает новые задачи. Чем больше накоплено данных и больше их используется для обучения параметров выбранной модели, тем лучше модель описывает реальность и может быть использована для более точного решения соответствующих задач;

- обучение у живых существ происходит методом проб и ошибок, корректировок своих представлений после получения результатов своих решений и действий, обобщением накопленного опыта;

Например, если вживить лягушке третий глаз (см. рис. ниже), то через некоторое время он научится распознавать объекты и сможет работать наравне с другими глазами [Constantine-Paton & Law, 2009]. Т.е. нервная система лягушки научится правильно интерпритировать сигналы, идущие с рецепторов данного глаза

## **2. Модели машинного обучения**

---

Рассмотрим основную модель данных, используемую при решении задач машинного обучения и основную модель принятия решения (решающей функции), используемую для обучения и прогноза.

### **2.1. Модель данных. Признаковое описание объектов. Виды признаков.**
---

#### **Признаковое описание объектов**

- определенное конечное или бесконечное множество однородных объектов (объекты реального мира/ситуации/процессы) $o^{(k)} \in O, k=1,..,K$ мы можем описать с помощью векторов значений признаков
$$\overline{x}^{(k)} = \left(x^{(k)}_1, x^{(k)}_2, ..., x^{(k)}_n \right), k=1,..,K$$ характеризующих эти объекты;

- для некоторых из этих объектов $o^{(j)}, j=1,..,J$ мы также имеем значение интересующего нас показателя (ответа)
$$y_j, j=1,..,J$$

Если мы объединим все описания объектов в одной матрице $X$, каждая строка которой содержит описание какого-то одного объекта, то мы получим **объектно-признаковую матрицу**. Число строк данной матрицы совпадает с числом описываемых матрицей объектов, а число столбцов совпадает с числом признаков, с помощью которых мы описываем эти объекты.

Все известные ответы мы также можем объединить в одной матрице-столбце $Y$.

Пара массивов $X, Y$ используется для обучения/построения модели, их связывающей.

#### ПРИМЕР 1. Процесс обучения
---
Если мы рассмматриваем процесс обучения, то объектом может быть обучающийся. У обучающегося может быть много различных характеристик - вес, рост, уровень знаний, характер, режим, ... Но, в зависимости от задачи исследования, от задачи, которую ставит перед собой аналитик, мы отбираем только те характеристики, которые влияют на нашу цель исследования.

Если нас интересует сможет ли обучающийся успешно пройти курс, то вряд ли нас будет интересовать его рост, вес, цвет глаз,... А вот его уровень знаний предмета, режим, в котором он привык обучаться, его стиль обучения - это нам будет интересно учесть.


### **2.2. Модель решающей функции. Модель машинного обучения.**
---
Мы бы хотели получить ответы - значение интересующего нас показателя $y$ (спрогнозировать оценку, оценить профиль поведения, спрогнозировать уровень коверсии курса, ...), когда будет предъявлен новый объект, который еще не встречался нам ранее.

#### **Опр. Решающая функция или модель машинного обучения**

> Пусть у нас есть множество описаний объектов $\bar x^{(k)}, k=1,..,K$ и могут быть известны cответствующие им (объектам) - метки $y_k, i=1,..,K$.

> Функция $f_m$, которая отображает с какой-то точностью множество описаний объектов $X$ в множество ответов $Y$ называется решающей функцией или моделью машинного обучения. Предполагается, что модель должна хорошо описывать некоторую неизвестную функцию f:

$$f: X \rightarrow Y$$

Для построения такой модели используется обучающая выборка

#### **Опр. Обучающая выборка**

> Под обучающей выборкой мы будем понимать некоторое множество L объектов, для которых известно их признаковое описание $\bar x^{(i)}, i=1,..,L$ и, возможно, их метки $y_i, i=1,..,L$  ($L$ - Learning, обучение).

> Эти описания мы будем объединять и обозначать матрицей $X^L$, каждая строка которой содержит описание какого-то объекта выборки.	Все известные ответы мы также будем обозначать $Y^L$.


Также нам будут нужны и другие выборки для анализа полученной модели - тестовая или контрольная. Их объектно-признаковую матрицу также будем обозначать буквами $X^T$ или $X^K$, а соответствующие метки  $Y^T$ или $Y^K$.

#### **Опр. Алгоритм машинного обучения**

> Алгоритм $a$, который по известной обучающей выборке $X^L (, Y^L)$ строит модель $f_m$ называется алгоритмом машинного обучения:

$$fm = a(X^L, Y^L) $$

Таким образом **решение задачи машинного обучения** - это построение алгоритма машинного обучения, который на основании обучающей выборки строит модель, которая затем используется для поиска ответов на новые вопросы.

**Задачи обучения с учителем (supervised learning)**

---
Самыми простыми задачами машинного обучения можно считать задачи "Обучение с учителем" или "Обучение по прецедентам" (supervised learning). Алгоритмы решения таких задач хорошо проработаны и, как правило, качество решения таких задач зависит только от качества данных. Подразумевается, что Учитель знает ответы на вопросы и обучает модель, показывая ей правильные ответы, чтобы она также могла находить эти ответы в будущем. Это очень похоже на обучение в школе. Когда ответы на учебные задачи заранее известны учителю.

- В этом классе задач известны (прецеденты) - описания объектов $X^{(i)}$ и ответы для них $y_i$ (метки) для достаточно большого набора объектов (обучающая выборка). В этом случае говорят, что имеются размеченные данные.

- мы обучаем алгоритмы на этой части объектов, для которых известны ответы и используем обученные алгоритмы для прогноза ответов на новых объектах, для которых неизвестны ответы (метки, классы).

- самой большой проблемой для решения этих задач - являются разметка большого количества данных.

### **3.1. Задача регрессионного анализа данных**

К задачам обучения с учителем можно отнести задачу регрессионного анализа данных. Когда нам нужно уметь прогнозировать какую-то величину (прибыль, итоговый балл на курсе) по известным входным параметрам (вложения, мероприятия, начальные результаты обучения, ...).

```
import numpy as np
import matplotlib.pyplot as plt   # библиотека построения графиков
plt.style.use('ggplot')   # устанавливаем стиль построения графиков
```

```
# Сгенерируем данные для аппроксимации
def f(x, eps=4.0, random_state = 20):
    np.random.seed(random_state)
    return 2 - 5*x + x**2 + (np.random.random(size= len(x))-0.5)*eps

x_list = np.arange(0.0, 7.0, 0.5); y_list = f(x_list);
x1_arr = np.arange(0.0, 6.1, 0.2)
plt.hlines(0, x_list.min(), x_list.max())
plt.vlines(0, y_list.min(), y_list.max())
plt.plot(x1_arr, f(x1_arr, eps=0), 'g--')
plt.scatter(x_list, y_list, s=100, marker='X');
```

Когда нам нужно уметь прогнозировать к какому классу относится данный объект (отличник, хорошист, троечник, ..) по известным входным параметрам (начальные результаты обучения, портфолио ...).

Специфика задачи классификации объектов заключается в том, что ответы в таких задачах являются бинарными или категориальными (номинальными или порядковыми) признаками.




```
np.random.seed(42)
# сгенерируем множество "нулей" как нормальное распределение N(m,s)
n0 = 10; m = np.array([0, 0]); s = 2
x0 = np.random.randn(n0, 2) * s #+ m
y0 = np.zeros(n0)

# сгенерируем множество "единиц" - точек "около" отрезка AB
n1 = 10; A = np.array([+2, 2]); B = np.array([+2, 6])
t1 = np.random.random((n1,1))
t2 = (np.random.random((n1,2))-0.5)*0.5
x1 = t1*A + (1-t1)*B + t2
y1 = np.ones(n1)
```

```
# оборазим графически расположение точек
plt.scatter(x0.T[0], x0.T[1], c='b', marker='o')
plt.scatter(x1.T[0], x1.T[1], c='r', marker='X');
```

```
# нарисуем границу a*x + b = 0
a = -3; b = 8
xline_0 = np.arange(-1, 3.1, 0.1)
xline_1 = a*xline_0 + b
plt.plot(xline_0, xline_1, 'g')

# и точки
plt.scatter(x0.T[0], x0.T[1], c='b', label='0')
plt.scatter(x1.T[0], x1.T[1], c='r', label='1')
plt.legend();
```

## **4. Задачи обучения без учителя (unsupervised learning)**
---

В задачах обучения без учителя нам не известны метки (классы), которым принадлежат объекты. В этом классе задач известны только описания объектов $X^{(i)}$ для достаточно большого набора объектов. Подразумевается, что модель должна сама разметить данные (найти ответы), чтобы свести задачу к задаче обучения с учителем.


Перечислим задачи анализа данных, с которыми обычно сталкивается дата-машинист, прежде чем сможет реализовать машинное обучение с учителем и решение реальной задачи

- предобработка данных: заполнение пропусков, приведение к удобному формату;
- добавление новых более удобных предикторов, меток объектов;
- графический и статистический анализ распределений объектов по имеющимся параметрам, по каждому параметру отдельно;
 выявить структуру данных:
	- снизить размерность данных с сохранением структуры,
	- графически проанализировать структуру данных,
	- кластеризовать объекты, чтобы выделить классы типовых, похожих друг на друга объектов, проанализировать их генезис, чтобы подумать как их можно использовать для решения поставленных задач.

К задачам анализа структуры данных, в частности, относятся:

- задача кластеризации; эту задачу приходится решать в том числе, чтобы разметить объекты;
- задача снижения размерности данных, задача многомерного шкалирования, факторный анализ;
- поиск правил, объясняющих причинную взаимосвязь между различными признаками в данных.

Рассмотрим виды и примеры таких задач

### **4.1. Задачи кластеризации объектов**
---

**Кластеризация** — объединение похожих объектов в группы, или	кластеры. Кластеризацию также называют кластерным анализом или анализом пространственной структуры объектов.

clusterization.png

Результаты кластеризации зависят от метода и от того, как оценивать схожесть или различие объектов, т.е. расстояние между ними. Чем дальше объекты друг от друга, тем меньше сходства и больше различия.

Кластеризация отличается от задачи классификации тем, что классы здесь заранее неизвестнеы и задачей кластеризации и является определить какие объекты в какие классы (кластеры) должны объединиться (см. рис. выше). Результат кластерного анализа и может служить разметкой данных для дальнейшего применения методов обучения с учителем.

Рассмотрим примеры поставновок задач кластеризации.

**Пример 1. Выявить кластеры типичных слушателей на онлайн-курсе**

Вы имеете большую статистику поведения слушателей на онлайн-курсе - как они проходят курс, как набирают баллы, в какой последовательности смотрят материалы. Вам хочется выделить обучающихся с похожим (типичным) поведением для того, чтобы выстроить для них эффектиное сопровождение.


Иногда важно не только выделить типичные объекты в кластеры, но и найти и выделить аномалии - нетипичные объекты, которые непохожи на других объектов. Можно считать эту задачу поиска аномалий - дополнением к задаче кластеризации. Некоторые методы кластеризации одновременно размечают и аномальные объекты.

**Пример 2. Поиск аномалий: обнаружение попыток подбора ответов в тестах**

Если тесты содержат вопросы с выбором ответов, то студент при большом количестве попыток может просто подбирать правильные ответы, даже не понимая сути. Тем не менее, эти ситуации можно распознать, анализируя какие ответы он выбирает в каждой попытке и есть ли прогресс в его попытках.

**Пример 4. Снижение размерности: кодирование многомерных объектов и визуальный анализ.**

Предположим, что мы имеем определенную статистику поведения сотрудников компании на обучающих корпоративных курсах. Вам хочется на основе этой статистики выделить сотрудников, которые не будут прогрессировать в работе. Но вы не знаете какая информация для этого будет полезна. На основе какой информации можно разделить сотрудников на кластеры, чтобы выделить интересующий нас кластер. В этом случае полезно взглянуть на пространственную структуру объектов на основе имеющейся информации и понять есть ли там интересующие нас кластеры. Для этого необходимо воспользоваться методом снижения размерности описания объектов, сохраняющем пространственную структуру объектов

### **4.3. Задачи поиска правил, описывающих закономерности в описании объектов**
---

Задача поиска правил тесно связана с задачей кластеризации или поиска типичных объектов. Дело в том, что типичные объекты схожи между собой тем, что они подчиняются одинаковым закономерностям. Например, если мы имеем описание студентов курса "Математика и Python" в виде векторов их оценок за всевозможные виды работ на курсе (балл.1, ..., балл.N), то разные кластеры обучающихся будут образованы в зависимости от того, что понимают/умеют и не понимают/не умеют слушатели курса.

Например, если слушатель не смог установить на компьютер Python и пакет блилиотек анаконду, то он не продвинется далее 2-й темы курса. Потому как далее требуется выполнять задания по программированию. Если он плохо знает дифференциальное исчисление, то ему не осилить разделы, связанные с оптимизацией и обучением моделей. Таким образом, по результатам обучения на курсе можно сформулировать правила вида:

ПРАВИЛО: **ЕСЛИ** балл.1 (установка Python) = 0 **ТО** балл.n (n>2) = 0

Таким образом ассоциативные правила представляют собой механизм нахождения логических закономерностей между признаками объектов.


**Пример 5. Поиск правил: выявление закономерностей покупок в супермаркетах**

Как вы думаете, зачем супермаркеты выдают нам карты лояльности, по которым мы получаем скидки и раздичные выгоды. Дело в том, что они собирают информацию о том, какие продукты вы покупаете, чтобы в дальнейшем размещать продукты так, чтобы вы, добираясь до своего продукта встретили на своем пути много других продуктов, которые такие как вы покупают вместе. Т.е. как бы подсказывают вам, что вам еще надо или желательно взять.

Фактически, управление супермаркетами, на основе статистики покупок строит аооциативные правила вида:

Если покупатель купил Товар-1 то он купит и Товар-2.

**Пример 6. Поиск правил: выявление закономерностей поведения жителей Умного Дома**

Умный Дом может собирать информацию о последовательности действий жителей (пришел домой, включил свет после захода солнца, попросил включить обогрев, когда температура опустилась ниже 18 градусов, ...) Мы хотим на основе данной информации построить правила, которые определяют некоторые привычки хозяев, чтобы в будущем им подсказывать или напоминать или автоматизировать эти действия

## **6. Фреймворки решения задач машинного обучения**
---

### **6.1. Схема решения задач машинного обучения**
---
Самыми простыми задачами машинного обучения можно считать задачи обучения с учителем. Алгоритмы решения таких задач хорошо проработаны и, как правило, качество решения таких задач зависит только от качества данных.

Схема решения задач машинного обучения состоит в организации двух циклов (см. рис. ниже);
- **цикл построения или обучения модели (build model)** состоит из шагов: собираем и загружаем данные (Collect and load data), готовим данные для обучения модели (Create pipeline), обучаем модель (Train model), оцениваем и улучшаем модель.
- **цикл применения модели (use model)** состоит из шагов: загружаем модель (load), делаем прогноз (Make predictions).

MLworkflow.png

Основной принцип решения задач обучения с учителем состоит в следующих шагах:

- формируется обучающая выборка $X^L$, $Y^L$, содержащая как можно больше разнообразных примеров объектов.
- выбирается структура модели - решающей функции $f_m(\bar p, x): X \rightarrow Y$; здесь $\bar p$ - параметры функции;
- выбирается лосс $\mathfrak L (\bar p, X^L, Y^L)$ - функция потерь, которая вычисляет усредненный штраф за расхождение между значением решающей функции $f_m(\bar p, x)$ и реальной меткой объекта $y$ для всех объектов с описанием (x, y) в обучающей выборке;
- выбирается и применяется алгоритм машинного обучения $a$, который оптимизирует параметры решающей функции $\bar p$ на обучающей выборке  $(X^L, Y^L)$ за счет минимизации функции потерь $a: \mathfrak L(\bar p, X^L, Y^L) \rightarrow min$.
- построенная модель $f_m(a, x)$ используется для прогноза на новых данных.

Python стал популярным благодаря наличию большого числа мощных библиотек реализации машинного обучения и конструирования искусственного интеллекта.

Одной из основных библиотек, которая реализует основной фреймворк машинного обучения, стала библиотека **scikit-learn** или **sklearn**.

Библиотека содержит много готовых наборов данных (datasets) и примеров применения методов машинного обучения к этим данных при решении перечисленных задач.

Обучение модели заключается в применении выбранного метода/алгоритма обучения. Каждый алгоритм в библиотеке реализуется своим классом. Поэтому для решения задачи необходимо:

- Создать объект данного класса: $$method = Method();$$
- создать и обучить модель - для этого обратиться к методу fit алгоритма/метода и передать ему обучающую выборку
$$model = method.fit(X^L, Y^L);$$

- после обучения модели мы можем воспользоваться ею для прогноза на новых данных:
$$Y = model.predict(X)$$

**ЗАКЛЮЧЕНИЕ**
---

- Основной принцип машинного обучения - это построение модели прнятия решения на основе данных; чем больше данных - тем точнее должна быть модель и лучше принимаемое решение;

- **решение задачи машинного обучения** означает построение алгоритма машинного обучения, который на основании обучающей выборки строит модель, которая затем используется для поиска ответов на новые вопросы;

- выделяют три основных типа задач машинного обучения в зависимости от того, известны или неизвестны ответы (прецеденты) в обучающей выборке:

- Схема решения задач машинного обучения состоит в организации двух циклов: *цикл обучения модели* и *цикл применения модели*.
