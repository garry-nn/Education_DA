## Анализ данных в современном мире
---

**Анализ данных** — это процесс очистки, изменения и обработки необработанных данных, а также извлечения полезной и актуальной информации, которая помогает предприятиям принимать обоснованные решения.

Анализ данных представляет собой процесс получения информации путем создания моделей и применения математического аппарата (точнее статистических законов) для поиска закономерностей.

Процедуры анализа данных помогают снизить риски, связанные с принятием решений, предоставляя полезную информацию и статистику, часто представленную в виде диаграмм, изображений, таблиц и графиков.


Образовательные данные могут быть представлены в разных форматах. Можно выделить следующие случаи анализируемх данных:

- **Большие данные**. Большие данные определяются  как огромный набор данных, который со временем продолжает расти в геометрической прогрессии. Четыре основные  характеристики больших данных  : объем, разнообразие, скорость и изменчивость. Объем описывает количество, скорость относится к скорости роста данных, а разнообразие указывает на разные источники данных. Достоверность говорит о качестве данных , определяя, представляют ли они ценность для бизнеса или нет.

- **Структурированные/неструктурированные данные**. Структурированные данные — это предопределенная модель данных, такая как традиционная база данных со столбцами строк. Неструктурированные данные поступают в формате, который не помещается в строки и столбцы, и могут включать в себя видео, фотографии, аудио, текст и многое другое. Сравнение  структурированных данных с неструктурированными  показывает, что структурированными данными легче управлять и анализировать.

- **Поток даннывх или данные в режиме реального времени**. Данные, которые представляются, как только они получены, известны как  данные в реальном времени. Этот тип данных полезен, когда решения требуют самой последней информации. Например, в режиме реального времени необходимо анализировать работу студентов с онлайн-курсом или результаты коворкинга.


### Место анализа данных в построении систем ИИ
---


Иерархию задач, которые необходимо решать при решении задач машинного обучения, созда
ния искусственного интеллекта хорошо отображает рис..

На рис. мы видим большое количество задач, которые надо решить, прежде чем удастся со-
здать "умную"программу, решающую задачу машинного обучения или построения искусственного
интеллекта. Эти задачи располагаются на самой вершине пирамиды задач.
Более того, ширина горизонтального сечения слоя пирамиды отражает тот факт, что трудо-
емкость задач внизу гораздо больше, чем трудоемкость задач вверху пирамиды.

Рассмотрим задачи, начиная с онования пирамиды.

Задача 1 (Collect data: сбор данных). В самом низу пирамиды находятся задачи по получе-
нию данных с раздичных сенсоров, логгирование процессов, сбор данных с различных внешних
источников и в том числе из опросов экпертов и пользователей систем.

image.png

Например, в рамках проекта "Умный Дом"требуется поставить в доме и вокруг него различные
сенсоры, датчики, вебкамеры, фиксировать звуки и команды людей, собирать отзывы о качестве
работы системы.

**Задача 2** (Move/Store: первоначальная очистка и накопление, хранение, структурирование).
После получения данных требуется эти данные где-то накапливать и хранить, предварительно
проверяя их надежность и отсеивая неинформативные данные.

Это задача связана с использованием облачных сервисов, на которых накапливаются данные.
При этом имеет смысл уже на этом этапе предобрабатывать сырые данные. Например, если весь
видеопоток со всех камер накапливать в течении нескольких суток, то это может занять сот-
ни терабайтов информации. И при этом событийных моментоа, ради которых ведется фиксация
может быть очень немного. Поэтому имеет смысл кодировать и запоминать только изменения в
картинках, а не сами картинки.

**Задача 3** (Explore/Transform: разведка/преобразование). на этом этапе происходит очистка
данных и их первичная предобработка перед последующим анализом. Поэтому на этом этапе могут
выявляться аномалии, которые могут внести серьезные искажения в результаты дальнейшего
анализа данных с целью их фильтрации или сглаживания.

Например, нет смысла хранить результаты сбоев видеокамер и других процессов, которые не
несут никакой информации. Недостоверные или искаженные данные. Резкие звуки или хлопки,
хрипы, которые могут сопровождать запись голоса нужно вырезать или уменьшить, отфильтро-
вать.

**Задача 4** (Aggregate/Label: агрегирование/разметка). на данном этапе происходит анализ и
подготовка данных для использвоания в алгоритмах машинного обучения. Данные анализируют-
ся, кластеризуются, агрегируются, размечаются

Например, чтобы обучить "Умный дом"правильно реагировать на команды хозяина, необхо-
димо собрать образцы голосовых команд, разметить их (т.е. указать, какая команда была произ-
несена), закодировать и проаназировать насколько хорошо произнесены команды и насколько их
можно отличить от других команд, или от команд, произнесенных другими людьми.

На данном этапе необходимо определить основные метрики, которые могут быть испльзуемы
для того, чтобы понять насколько хорошо решение той или иной задачи. Т.е. выделить числовые
критерии качества решения задач.

Основные задачи аналитика - это задачи 3 и 4. Задачи сбора и хранения даннных решаются на образовательных порталах, на современных образовательных платформах.

## Анализ данных в образовании
---
Анализ данных в образовании — это использование машинного обучения, статистики и Data Mining для обработки информации из учебных платформ (курсы, электронные журналы) с целью персонализации обучения, прогнозирования успеваемости и оптимизации учебных программ. Он помогает выявлять пробелы в знаниях, снижать отсев студентов и повышать эффективность методик обучения.

Особенно остро необходимость в анализе данных появилась после массового применения онлайн-курсов, на которых иногда обучаются одновременно десятки тысяч пользователей.

### Основные направления применения анализа данных в образовании:

- Персонализация обучения (адаптивное обучение): Автоматическая корректировка учебных программ под уникальные потребности, темп и способности каждого студента.

- Прогнозирование успеваемости: Выявление учащихся «в зоне риска» (с риском отчисления или низкой успеваемости) на основе анализа их поведения, посещаемости и оценок, что позволяет педагогам вовремя вмешаться.

- Анализ образовательного контента: Оценка эффективности использования мультимедийных ресурсов, электронных учебников и онлайн-курсов.

- Оптимизация процессов и бюджета: Улучшение административного планирования, управление расходами и разработка новых образовательных продуктов.


**Типы данных, подлежащие анализу**

- Персональные данные: Базовая информация об учащемся.

- Данные об обучении: Взаимодействие с контентом (время просмотра, результаты тестов, скорость прохождения модулей).

- Административные данные: Посещаемость, пропуски, зачеты.

**Инструменты и технологии:**

Аналитика часто реализуется с помощью языка Python (библиотеки, Jupyter Notebook, Google Colab), специализированных инструментов анализа даннных, а также платформ хранения и обработки данных, таких как Hadoop.

## Этапы анализа данных. Задачи.

Этапы анализа данных включают в себя сбор всей информации, ее обработку, изучение данных и их использование для поиска закономерностей и других идей. Процесс состоит из:

- **Сбор требований к данным:** спросите себя, почему вы проводите этот анализ, какой тип анализа данных вы хотите использовать и какие данные вы планируете анализировать.

- **Сбор данных:** руководствуясь определенными вами требованиями, пришло время собирать данные из ваших источников. Источники включают тематические исследования, опросы, интервью, анкеты, прямое наблюдение и фокус-группы. Обязательно систематизируйте собранные данные для анализа.

- **Очистка данных:** не все данные, которые вы собираете, будут полезны, поэтому пришло время их очистить. В этом процессе вы удаляете пробелы, повторяющиеся записи и основные ошибки. Очистка данных обязательна перед отправкой информации на анализ.

- **Анализ данных:** здесь вы используете программное обеспечение для анализа данных и другие инструменты, которые помогут вам интерпретировать и понимать данные и делать выводы. Инструменты анализа данных включают Excel, Python , R, Looker, Rapid Miner, Chartio, Metabase, Redash и Microsoft Power BI.

- **Интерпретация данных:** теперь, когда у вас есть результаты, вам нужно их интерпретировать и предложить наилучший план действий на основе ваших выводов.

- **Визуализация данных:** визуализация данных — это причудливый способ сказать: «графически покажите свою информацию так, чтобы люди могли ее прочитать и понять». Вы можете использовать диаграммы, графики, карты, маркеры или множество других методов. Визуализация помогает получать ценные сведения, помогая сравнивать наборы данных и наблюдать за взаимосвязями.


### Виды анализа данных

Сегодня доступно полдюжины популярных типов анализа данных, обычно используемых в мире технологий и бизнеса. Они есть:

- **Описательный:** описательный анализ работает отвечает на вопрос: «Что происходит?». Он описывает происходящий процесс, иллюстрирует средние значения и отклонения в непрерывных данных, а также проценты и частоты в категориальных данных.

Данный анализ используется для решения задач мониторинга учеюного процесса. Рассчитываются средние баллы, процент (НЕ)успевающих в группе по результатам аттестации, процент посещения занятий, ...

- **Диагностический анализ:** диагностический анализ отвечает на вопрос: «Почему это произошло?» Используя информацию, полученную в результате статистического анализа (подробнее об этом позже!), аналитики используют диагностический анализ для выявления закономерностей в данных. В идеале аналитики находят аналогичные модели, существовавшие в прошлом, и, следовательно, используют эти решения для решения текущих проблем.

Например, на заре внедрения онлайн-курсов и онлайн-обучения (в 2012-2016 гг) многие разработчики онлайн-курсов обращались к аналитикам с вопросом "Почему так мало обучающихся завершают успешно их онлайн-курс?", "В чем состоит камень преткновения?" "Почему они бросают его на полпути?". "Как уулучшить курс?", "Что вообще происходит на курсе?". Примеры ОК УрФУ (Open EDU), ТГУ(Coursera), МФТИ(Coursera), ВШЭ (Coursera).

- **Предиктивный анализ:** предиктивный анализ отвечает на вопрос: «Что, скорее всего, произойдет?» Используя закономерности, обнаруженные в старых данных, а также в текущих событиях, аналитики предсказывают будущие события. Хотя не существует такой вещи, как 100-процентно точный прогноз, шансы улучшаются, если у аналитиков есть много подробной информации и дисциплина для ее тщательного исследования.

Например, в ЦЭО Волгатеха анализировали результаты обучения на эл. курсах после 1-й аттестиации (8-я неделя обучения, экватор) и прогнозировали итоговые результаты (процент успевающих после сессии).

- **Статистический анализ:** cтатистический анализ не только углубляет ответ на вопрос: «Что произошло?» и "почему произошло", он анализирует взаимосвязи. Таким образом, это важная составляющая и описательного и предиктивного и диагностического анализа. В статистическом анализе выделяют несколько подкатегорий: корреляционный анализ, регресиионный анализ, проверка гипотез.

- **Анализ текста:** также называемый « интеллектуальным анализом данных », анализ текста использует базы данных и инструменты интеллектуального анализа данных для обнаружения шаблонов, находящихся в больших наборах данных. Он преобразует необработанные данные в полезную бизнес-информацию. Анализ текста, пожалуй, самый простой и прямой метод анализа данных.

Данный анализ может быть использован для анализа и оценки ответов - эссе студентов, их письменных работ. Можно сравнивать ответы с шаблонами ответов или заполнять чек-лист наличия в ответе каких-то важных деталей.

- **Предписывающий анализ:** используйте все типы анализа, сравните результаты анализа с нормативными показателями, сделайте выводы, рекомендации и вы получите предписывающий анализ.

Такой анализ очень часто полезен для создателей онлайн-курсов, для создателей тестов. Диагностика показывает в чем проблема тестового задания (слабая дифференциация, слишком легкое задание, ...)  и рекомендательная система предписывает как это можно исправить.

## Формат хранения и представления данных для анализа.
---
Большинство данных хранятся в реляционных базах данных в виде связанных таблиц.
Мржно получить одну таблицу данных или связать несколько таблиц в одну таблицу в формате xls, xlsx или в формате csv.


### Формат CSV
---
В формате csv данные хранятся в текстовом виде и могут легко просматриваться человеком. Пример таких данных можено видеть на сайте открытых данных Минобра - результаты ЕГЭ (https://opendata.edu.gov.ru/opendata/7710539135-ege):

```
"region","count_vyp","count_participant","count_100ball"
"Республика Адыгея","2833","3232","5"
"Республика Башкортостан","25731","28138","27"
"Республика Бурятия","6890","7547","15"
"Республика Алтай","1694","2180","2"
...

```

1-я строка такого файла может содержать названия столбцов (параметров).
Каждая следующая строка содержит занчения этих параметров для какого-либо объекта.

## Пакеты Python для анализа данных

Чтобы провести анализ данных в Python за минимальное время, мы должны сначала познакомиться с основными пакетами (библиотеки).

Библиотека Python - это набор функций и методы, которые позволяют выполнять множество действий без написания кода. Библиотеки обычно содержат встроенные модули, содержащие разные функции, которые можно использовать, вызвав определенный метод библиотеки.

Все библиотеки анализа данных Python делятся на четыре группы, в которых представлены библиотеки, позволяющие решать задачи разной уровни сложности:

### Бюблиотеки обработки данных

Первая группа называется научными **вычислительными библиотеками**. Сюда относится библиотека **Pandas**. Она является одной из частоиспользуемых библиотек при работе с большими данными. Библиотека Pandas предполагает работу со структурами данных и предоставляет инструментарий эффективного манипулирования данными и их анализа.  

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/pandas_logo.png" width="200"/>


Pandas отлично проявляет себя в сферах статистики, анализа данных, экономики, финансов. **Основная задача библиотеки - очистка и первичная оценка данных по общим показателям**. Инструментом библиотеки является двухмерная таблица, состоящая из меток столбцов и строк и называя **фреймом данных (DataFrame)**. Имеется поддержка объекта DataFrame, предназначенного для работы с индексированными массивами и позволяющий работать с различными форматами данных: начиная от CSV до TSV.

Библиотека уверенно справляется с обработкой больших объёмов данных и при необходимости использует возможности NumPy и SciPy. Впрочем именно взаимодополнение этих пакетов и привело к тому, что чаще всего их используют вместе. Pandas позволяет:

- Обменивать данные между структурами в памяти и файлами различных форматов;
- Переформатировать наборы данных (даже создавать сводные таблицы);
- Выполнять трёхэтапные операции типа «разделение, изменение, объединение»;
- Работать с временными рядами: формирование временных периодов и изменение интервалов.

**Библиотека NumPy** - это один из основных пакетов программного обеспечения научной библиотеки приложений для обработки больших многомерных массивов и матриц, большой набор расширенных математических функций и методов реализации которых, позволяет этим объектам выполнять операции. Лучшая и самая важная особенность Numpy — **интерфейс массива**.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/numpy_logo.png" width="200"/>

Его можно расширить до объектов для матриц и с небольшим изменением кода разработчики быстрее выполняют обработку. Этот интерфейс можно использовать для выражения изображений, звуковых волн и других двоичных необработанных потоков в виде массива действительных чисел в n-мерном формате. В общем, библиотека полезна для full-stack разработчиков, занимающихся задачами машинного обучения.

В NumPy (Numerical Python) встроено множество вычислительных механизмов. К таким особенностям можно отнести:

- Возможность использования и выполнения простых и достаточно сложных научных расчётов;
- Поддержка многомерных массивов;
- Множество встроенных методов для выполнения различных вычислений с большими и многомерными массивами данных;
- Возможность преобразования данных;
- Поддержка не только с числовых, но и других типов данных.

**Библиотека SciPy** - эта библиотека построена на базе NumPy. SciPy отлично подходит для выполнения статистических расчётов и поддерживает те же виды данных, что и библиотека NumPy. Можно сказать, что SciPy - это как "матрешка в матрешке": набор подпакетов, в которых реализованы различные вычислительные механизмы.

<img src="https://github.com/SerjiEvg/data-analysis/raw/main/images/scipy_logo.png" width="200"/>


SciPy предоставляет все **эффективные численные процедуры**, такие как оптимизация, численное интегрирование и многие другие, с использованием специальных подмодулей. Например, быстрое преобразование Фурье, обработку изображений, решение дифференциальных уравнений, механизмы линейной алгебры и так далее. Благодаря дополнительным пакетам SciPy может решать задачи, которые "не по зубам" NumPy.

SciPy использует массивы NumPy в качестве базовой структуры данных и поставляется с модулями для различных и частоиспользуемых задач в научном программировании.

**Библиотеки для визуализации данных** - лучший способ визуально показать содержательные результаты анализа. Эти библиотеки позволяют создавать графики, диаграммы и различные карты.
